```python
import pandas as pd
import unicodedata
import emoji
import re
from unidecode import unidecode
import math
from decimal import Decimal, ROUND_DOWN
from ast import literal_eval
```

```python
#fonction qui nettoye le texte en enlevant les accents et en mettant le texte en minuscule
def clean_text(text):
    return unidecode(text).lower()
```

## Picture

```python
# fonction qui retourne si l'annonceur à une photo de profil ou non
def profil_picture(row):
    if row['photo']!='Null':
        return 1
    else:
        return 0
```

## Description user

```python
# fonction qui retourne si l'annonceur à une description ou non
def description_user_full(row):
    if pd.isna(row['description_user']):
        return 0
    else:
        return 1

# fonction qui retourne si l'annonceur semble avoir un compte de revente de chaussures
def sell_sneaker_account(row):
    voc_snearker_account=['sneakers','sneakers','chaussures','chaussure','resseler','streetwear','vintage']
    
    # Vérifier si la valeur n'est pas NaN
    if pd.notna(row['description_user']):
        for word in voc_snearker_account:
            if clean_text(word) in clean_text(row['description_user']):
                return 1
    return 0
```

## Price

```python
# fonction qui retourne si un prix est décimal mais différent de 0.9, 0.99 et 0.5
def price_decimal(row):
    decimal_part = Decimal(str(row['price'])) % 1
    if decimal_part not in [Decimal('0.5'), Decimal('0.99'), Decimal('0.9'),Decimal('0.0')]:
        return 1
    else:
        return 0
```

## Description advert

```python
# fonction qui retourne si une annonce à le même texte pour le titre et la description
def identic_title_description(row):
    if row['title_advert']==row['description_advert']:
        return 1
    else:
        return 0

# fonction qui retourne si une description est courte
def short_description(row):
    if len(row['description_advert'].split())<8:
        return 1
    elif len(row['description_advert'].split())<=10 and '\n' not in row['description_advert']:
        return 1
    else:
        return 0

# fonction qui retroune si la description semble secure grace à des mots de confiance
def secure_description(row):
    voc_ressell=['remise en main propre','preuves d’authenticité','remise en mains propres',
    'preuve d\'achat','facture','magasin','double box','carton de protection','ticket de caisse']
    for word in voc_ressell:
        if clean_text(word) in clean_text(row['description_advert']):
            return 1
    return 0

# fonction qui retroune si la description comporte un/des emoji(s)
def has_emoji_description(row):
    for char in row['description_advert']:
        if emoji.emoji_count(char):
            return 1
    return 0
```
