```python
def is_fraudulent_username(username):
    # Vérifier s'il contient 3 consonnes ou 3 voyelles consécutives
    if re.search(r'[aeiou]{3,}|[bcdfghjklmnpqrstvwxyz]{3,}', username, re.IGNORECASE):
        print(username,' frauduleux')
        return True
    # Vérifier s'il contient des chiffres (garder uniquement ceux en fin de pseudo)
    digits_in_username = re.findall(r'[a-zA-Z]+(\d+)[a-zA-Z]+', username)
    if bool(digits_in_username)==True:
        print(username,' frauduleux')
        return True

    # Vérifier s'il contient des caractères spéciaux
    special_characters = re.compile(r'[!@#$%^&*(),?":{}|<>]')
    if special_characters.search(username):
        print(username,' frauduleux')
        return True

    # Si toutes les conditions sont satisfaites, le pseudo est considéré comme sûr
    print(username,' detente')
    return False
```
